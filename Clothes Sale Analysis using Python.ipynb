### Data Cleaning

```python
# Replace ;,; with +
daily_sales_comma = daily_sales.replace(";,;", "+")
```

```python
# First 100 characters
daily_sales_comma[:100]
```

```python
# Split with ,
daily_transactions = daily_sales_comma.split(",")
daily_transactions[:10]
```

```python
# Split by +
daily_transactions_split = []
for transaction in daily_transactions:
    daily_transactions_split.append(transaction.split("+"))
print(daily_transactions_split[0:10])
```

```python
# Extract Information
transactions_clean = []
for transaction in daily_transactions_split:
    transaction_clean = []
    for data_point in transaction:
        transaction_clean.append(data_point.replace(" \n", "").strip(" "))
    transactions_clean.append(transaction_clean)
print(transactions_clean[0:10])
```

```python
# Extract customers, sales and thread_sold
customers = []
sales = []
thread_sold = []

for transaction in transactions_clean:
    customers.append(transaction[0])
    sales.append(transaction[1])
    thread_sold.append(transaction[2])
```

```python
# DataFrame
df = pd.DataFrame({"Customers": customers,
                  "Sales": sales,
                  "Thread Sold": thread_sold})
df
```

```python
# Remove $ sign
df["Sales"] = df["Sales"].str.lstrip("$").astype("float")
```

```python
def Class(x):
    if x >= 0.0 and x <= 10.0:
        return "Economic"
    elif x > 10.0 and x <= 20.0:
        return "Standard"
    elif x > 20.0 and x <= 31.0:
        return "Expensive"
```

```python
# Rename Column
df = df.rename(columns = {"Sales": "Sales $"})
df["Class"] = df["Sales $"].apply(Class)
df.head()
```

### Deep Analysis

```python
# Frequent Customers
c = df["Thread Sold"].value_counts().to_frame().reset_index()
c.columns = ["Color", "Thread Sold"]
c
```

```python
dfc = df["Class"].value_counts().to_frame().reset_index()
dfc.columns = ["Class", "Clothes Sold"]
dfc
```

```python
plt.figure(figsize = (18, 7))
graph = sns.barplot(x = "Class", y = "Clothes Sold", data = dfc, palette = "spring")
for p in graph.patches:
        graph.annotate('{:.01f}'.format(p.get_height()), 
                      (p.get_x()+0.41, p.get_height()),
                       ha='center', va='bottom',color= 'black', size = 15)
plt.title("Clothes Sold by Class", size = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel(None)
plt.ylabel(None)
plt.show()
```

**Min, Mean and Max Sales by Class**

```python
dfgc = df.groupby("Class")["Sales $"].agg(["min", "max", "mean"]).reset_index()
dfgc.columns = ["Class", "Min Sale $", "Max Sales $", "Avg Sale $"]
dfgc
```
